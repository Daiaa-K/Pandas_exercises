### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
data = pd.read_csv(r"https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv")
```

```python
data.head()
```

```python
data.info()
```

### Step 4. Set PassengerId as the index 

```python
data = data.set_index("PassengerId")
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(data.Sex.value_counts(),labels=data.Sex.unique(),autopct='%1.2f%%')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(data=data,x="Fare",y="Age",hue="Sex")
```

### Step 7. How many people survived?

```python
len(data.query('Survived == 1 '))
```

```python
sns.countplot(data=data,x="Survived")
```

### Step 8. Create a histogram with the Fare payed

```python
sns.histplot(data=data,x="Fare",kde=True)
```
